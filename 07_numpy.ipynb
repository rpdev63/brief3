## Challenge Numpy : Modéliser la performance d'un marathonien à l'aide d'une régression

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv(r'ressources/marathon.txt', delimiter="\t", names=["City", "Year", "Time","Seconds"])
```

**Ajout des colonnes isParis et isBerlin au dataframe**

```python
isParis =[]
isBerlin = []
for City in df['City'] :
    if City == "PARIS" :
        isParis.append(1)
        isBerlin.append(0)
    elif City == "BERLIN" :
        isParis.append(0)
        isBerlin.append(1)
    else :
        isParis.append(0)
        isBerlin.append(0)

df['isParis'] = isParis
df['isBerlin'] = isBerlin

df.head(5)
```

**On essaie de trouver A dans l équation suivante Y = XA**

```python
#Y = XA
#A* = (X'X)^-1 X'Y
X=[]
Y=[]

for i in df.index:
    inY = [df['Seconds'][i]]
    Y.append(inY)
    inX = [ df['Year'][i],df['isParis'][i],df['isBerlin'][i] ]
    X.append(inX)
    
X = np.matrix(X)

Y = np.matrix(Y)

X
```

```python
Y
```

```python
Xt = X.T
Xt
```

```python
A = (Xt * X).I * Xt * Y
print(A)
```

```python
print("Les coefficients sont les suivants : \na0 = {} \na1 = {} \na2 = {} ".format( A[0,0], A[1,0], A[2,0]))
```
